```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.svm import SVR
%matplotlib inline
```

```python
# Load dataset
df = pd.read_csv('hdd_prices_clean.csv')
data = np.matrix(df)
```

## Linear Regression

```python
# Split into train test
capacity = data[:, 1]
price = data[:, 2]
rng_state = np.random.get_state()
np.random.shuffle(capacity)
np.random.set_state(rng_state)
np.random.shuffle(price)
train_lines = int(0.8*len(capacity))
print train_lines
X_train = capacity[:train_lines]
X_test = capacity[train_lines:]
Y_train = price[:train_lines]
Y_test = price[train_lines:]
```

```python
# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, Y_train)
```

```python
# Explained variance score: 1 is perfect prediction
print 'Variance score: %.2f' % regr.score(X_test, Y_test)
```

```python
# Plot outputs
plt.scatter(df.capacity, df.price, color='black', label='data')
plt.plot(X_train, regr.predict(X_train), color='blue',
         linewidth=3, label='Linear regression')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

## Support Vector Machine

```python
# Start and fit SVM for regression using different kernels
X = X_train
y = np.array(Y_train).ravel()

svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
y_rbf = svr_rbf.fit(X, y).predict(X)
print "Score: %s" % (svr_rbf.fit(X, y).score(X_test, np.array(Y_test).ravel()))
```

```python
# look at resulst
plt.scatter(df.capacity, df.price, c='k', label='data')
plt.hold('on')
plt.plot(X, y_rbf, c='g', label='RBF model')
plt.plot(X, regr.predict(X_train), c='r', label='Linear regression')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
